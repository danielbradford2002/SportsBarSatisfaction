```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload your file in Google Colab
from google.colab import files
uploaded = files.upload()

# Load the dataset (update the filename to match the uploaded file)
data = pd.read_excel('Final_data (1).xlsx')  # Use the correct uploaded filename

# 1. Distribution of Ratings During Event vs Off-Season
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='event_season', y='stars')
plt.title("Distribution of Ratings: Event vs. Off-Season")
plt.xticks([0, 1], ['Off-Season', 'Event'])
plt.ylabel("Ratings")
plt.xlabel("Season Type")
plt.show()

# 2. Average Review Counts During Event vs Off-Season
plt.figure(figsize=(8, 6))
sns.barplot(x='event_season', y='review_count', data=data, ci=None)
plt.title("Average Review Counts: Event vs. Off-Season")
plt.xticks([0, 1], ['Off-Season', 'Event'])
plt.ylabel("Average Review Counts")
plt.xlabel("Season Type")
plt.show()
```

```python
# Calculate statistics for ratings during Event and Off-Season
event_stats = data[data['event_season'] == 1]['stars'].describe()
off_season_stats = data[data['event_season'] == 0]['stars'].describe()

print("Event Season Ratings Statistics:")
print(event_stats)

print("\nOff-Season Ratings Statistics:")
print(off_season_stats)

# Box Plot with Adjusted Range
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='event_season', y='stars')
plt.ylim(1.0, 5.5)  # Adjust the y-axis range to better visualize data
plt.title("Distribution of Ratings: Event vs. Off-Season")
plt.xticks([0, 1], ['Off-Season', 'Event'])
plt.ylabel("Ratings")
plt.xlabel("Season Type")
plt.show()
```
